```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
```

### Load the dataset and preview 

```python
data = pd.read_csv(r'C:\Users\user\Documents\Applied Machine learning\Percept1.csv', sep=";", header=None)
```

```python
data, data.columns, data.info(), data.describe()
```

```python
X1 = data.iloc[:, 0:2]
print(X1)
```

### MinMax Normalization

```python
scaleMinMax = MinMaxScaler()
# (x- min) / (max - min)
X1_scaled = scaleMinMax.fit_transform(X1)
X1_scaled = pd.DataFrame(X1_scaled, columns=X1.columns)
print(X1_scaled)
```

```python
X1_scaled.describe()
```

```python
y = data.iloc[:, 2]
print(y)
```

### Splitting the dataset into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X1_scaled, y, test_size=0.2, random_state=42)
```

```python
X_train.describe()
```

```python
X_test.describe()
```

### Create perceptron model

```python
model = Perceptron()
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

### Evalution

```python
accuracy = accuracy_score(y_test, y_pred)

print("Prediction", y_pred)
print("Accuracy", accuracy)
```
